```python
from sklearn import datasets

import matplotlib.pyplot as plt
from cv2 import cv2 as cv
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn import metrics,svm
from sklearn.ensemble import RandomForestClassifier as RF
#Load the digits dataset
digits = datasets.load_digits()

x_train, x_test, y_train, y_test = train_test_split(digits.data,digits.target,test_size = 0.25)
# if we use digits.imgaes, there will be three dims as digits.images.shape = [1797,8,8]


neigh = KNN(n_neighbors = 3)
neigh.fit(x_train,y_train)
KNN_predict = neigh.predict(x_test)
score = neigh.score(x_test,y_test) # score = 0.991
print(score)
# if n_nighbors=5, it would cause some mistakes, very few mistakes will happen if n_neighbors=3


svm_model = svm.SVC() # use default
svm_model.fit(x_train,y_train)
svm_predict = svm_model.predict(x_test)
score = svm_model.score(x_test,y_test) # score = 0.991
print(score)

rf_model = RF()
rf_model.fit(x_train,y_train)
rf_predict = rf_model.predict(x_test)
score = rf_model.score(x_test,y_test) # score = 0.98
print(score)


print(digits.data.shape)
print(x_train.shape)

#KNN
print('------KNN------\n%s'%metrics.classification_report(y_test, KNN_predict))
disp_KNN = metrics.plot_confusion_matrix(neigh, x_test, y_test)
disp_KNN.figure_.suptitle("KNN Confusion Matrix")
print("Confusion matrix:\n%s" % disp_KNN.confusion_matrix)

#SVM
print('------SVM------\n%s'%metrics.classification_report(y_test, svm_predict))
disp_SVM = metrics.plot_confusion_matrix(svm_model, x_test, y_test)
disp_SVM.figure_.suptitle("SVM Confusion Matrix")
print("Confusion matrix:\n%s" % disp_SVM.confusion_matrix)

#RF
print('------RF------\n%s'%metrics.classification_report(y_test, rf_predict))
disp_RF = metrics.plot_confusion_matrix(rf_model, x_test, y_test)
disp_RF.figure_.suptitle("RF Confusion Matrix")
# print("Confusion matrix:\n%s" % disp_RF.confusion_matrix)

# from the accuracy, we will print KNN and SVM's confusion matrix





# cv.imshow('gg',digits.images[0])
# cv.waitKey(0)
```
